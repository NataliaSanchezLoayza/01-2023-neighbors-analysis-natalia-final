```python
brookhaven_data = pd.read_csv("../output/msg_extracts.csv")
brookhaven_data.head()
```

```python
brookhaven_data["body_title"]
```

```python
brookhaven_data['package_car'] = brookhaven_data['body_title'].apply(lambda x: 'package' in str(x).lower() or 'car' in str(x).lower())
brookhaven_data.head()
```

```python
brookhaven_data['package_car'].value_counts()
```

```python
brookhaven_car_package = brookhaven_data['package_car'].value_counts()
```

```python
brookhaven_car_package.to_csv("../output/car_package-Sanchez.csv")
```
